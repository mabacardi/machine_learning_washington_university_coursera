```python
import pandas as pd
```

```python
df=pd.read_csv("/content/home_data.csv",error_bad_lines=False, engine="python",  keep_default_na=False, delimiter=",")
names= df.columns #.str.split(',').tolist()
df= df.iloc[:,0].str.split(',', expand=True)
#print(len(names), len(df.columns))
df.set_axis(names, axis=1, inplace=True)
df
```

```python
# remove semicolons
df=df.replace('"','', regex=True)
df.head()
```

```python
# transform every column into the desired one

# create a list with the columns of the df
names=list(df.columns)
# remove the column with a different type (we could have done the same with integer columns)
names.remove("date")
print(names)
# iterate over the list without date
for n in names:
  df[n]=df[n].astype(str).astype(float)
# and now datetime works differently, so we change it separately
df.date=pd.to_datetime(df.date)
df.head() # display results
```
